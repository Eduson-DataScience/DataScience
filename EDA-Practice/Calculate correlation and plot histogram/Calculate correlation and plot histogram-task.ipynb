# Рассчитайте корреляцию и постройте гистограмму
## Условие задачи

Вы работаете в компании, занимающейся анализом данных, и вам поставлена задача исследовать взаимосвязь между двумя переменными в наборе данных: цена товара и объем продаж. Ваша цель — рассчитать корреляцию между этими переменными и построить диаграмму рассеяния для визуализации результатов.

Алгоритм решения:

1. Загрузите данные из файла формата CSV.
2. Проведите предварительный анализ данных: изучите структуру набора данных, определите переменные, которые вы хотите исследовать.
3. Выделите две переменные, между которыми вы хотите рассчитать корреляцию.
4. Проведите расчет корреляции между переменными с использованием подходящего статистического метода, проверьте значимость полученного коэффициента корреляции.
5. Постройте диаграмму рассеяния, отображающую взаимосвязь между двумя переменными.
6. Проанализируйте результаты и сделайте выводы о степени взаимосвязи между переменными.

**Входные данные (data.csv):**

[data.csv](https://github.com/Eduson-DataScience/EDA-Practice/blob/main/Calculate%20correlation%20and%20plot%20histogram/data.csv)

## Подсказки

- Используйте библиотеку Pandas для загрузки данных из файла CSV и работы с ними.
- Для расчета корреляции между переменными можно воспользоваться методом `**pearsonr**`.
- Для построения диаграммы рассеяния используйте функцию **`scatter()`** из библиотеки Matplotlib.
- Помните, что перед построением диаграммы рассеяния данные должны быть числовыми, поэтому, при необходимости, преобразуйте их в числовой формат.
- Анализируйте результаты, обращайте внимание на значение корреляции и визуализацию, чтобы сделать выводы о взаимосвязи между переменными.
